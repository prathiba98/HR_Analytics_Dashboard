```python
import pandas as pd
df = pd.read_csv("hr_cleaned.csv")
```

```python
df.head(2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='Attrition',color='teal')
plt.title("Overall Attrition Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Department', hue='Attrition',palette='dark:teal')
plt.title("Attrition by Department")
plt.show()
```

```python
sns.countplot(data=df, x='Gender', hue='Attrition',palette='dark:teal')
plt.title("Attrition by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(
    data=df,
    x='Age',
    hue='Attrition',
    bins=30,
    kde=True,
    palette=['black', 'teal']
)
plt.title("Age Distribution vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome',color='teal')
plt.title("Monthly Income vs Attrition")
plt.show()
```

```python
sns.countplot(data=df, x='JobSatisfaction', hue='Attrition',palette='dark:teal')
plt.title("Job Satisfaction vs Attrition")
plt.show()
```

```python
sns.countplot(data=df, x='WorkLifeBalance', hue='Attrition',palette='dark:teal')
plt.title("Work-Life Balance vs Attrition")
plt.show()
```

```python
sns.countplot(data=df, x='OverTime', hue='Attrition',palette='dark:teal')
plt.title("OverTime vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='YearsAtCompany', hue='Attrition', bins=20,palette=['black','teal'])
plt.title("Years at Company vs Attrition")
plt.show()
```

* Research and Development department has higher attrition.

* Younger employees are more likely to leave.

* Lower salary correlates with higher attrition.

* Overtime is a major attrition trigger.

* Most attrition occurs in early tenure.
